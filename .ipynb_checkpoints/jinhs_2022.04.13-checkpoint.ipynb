```python
import numpy as np
x = np.array([[1,2,3], [2,3,4], [3,4,5], [4,5,6]]) #입력 
y = np.array([4,5,6,7]) #예측 (바로 다음 수치값 예측). 실제 텍스트 데이터 처럼

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM #Dense는 바닐라, LSTM은 좀따 설명

model = Sequential()
model.add(LSTM(10, activation='relu', input_shape=(3,1))) # [1,2,3] input이 column vector로 들어감
model.add(Dense(5))
model.add(Dense(1)) # output 개수는 1개
model.summary()
```

```python
print(x.shape)
print(x)
x = x.reshape(x.shape[0], x.shape[1], 1) #이렇게 해야 각 cell에 scalar가 아닌 vector가 들어가서 연산이 됨 
print(x.shape) # (4,3,1) 최소 1명으로 이루어진 3묶음이 4묶음이 있다.
x
```

```python
model.compile(optimizer='adam', loss='mse')
model.fit(x,y, epochs=500)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import utils

data = list("hihello")
print(data)
token = Tokenizer()
token.fit_on_texts(data)
print(token.word_index) # 왜 index가 0이 아니고 1부터냐? 0번 인덱스는 의미없는 값을 표현하기 위해
x = token.texts_to_sequences(data)
print(x)
padded_x = pad_sequences(x, 1) # 데이터 1개 짜리 sequnces 데이터를 만들어라. 
# 원래는 문장 길이가 다르기 때문에, 길이를 맞추는 작업을 하는게 pad_sequecnes다. 근데 여기선 1로 똑같음.
print(padded_x)
one_hot = utils.to_categorical(padded_x)
print(one_hot) # 여기서 [0,0,0,0,0,0] 은 아무 의미가 없는 데이터를 표현하기 위해서
```

```python
from tensorflow.keras.layers import Input # 지금까지는 input 레이어가 없었다.
# 그냥 첫번째 hidden layer에 input dim을 설정헀는데 이제 명시적으로 한다.
# 이제부터 하려는건 hihell 을 입력으로 넣어서
# ihello 를 예측하려는 것
# 6개의 차원으로 6개의 차원을 출력하는걸 만들어야 함
# h를 넣어서 i로 나와야 하니까

model = Sequential()
model.add(Input(shape=(6, 6))) 
model.add(LSTM(12, return_sequences=True)) # true: 모든 cell의 출력을 다 사용하겠다
model.add(Dense(6, activation='sigmoid')) # output 개수가 6개, 각각의 값이 0, 1이니까 sigmoid.

model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary() # 6개의 cell이 각각 12개씩 내보낸다
```

```python
docs = ["나는 당신을 사랑합니다.", #input 1
        "나는 당신을 미워합니다.", #input 2
        "I love you.", #output 1
        "I hate you."] #output 2

# 우선 수치화를 해야하므로 tokenize를 하자
token = Tokenizer()
token.fit_on_texts(docs)
print(token.word_index)
```

```python
x = token.texts_to_sequences(docs)
print(x)
```

```python
padded_x = pad_sequences(x, 3)
print(padded_x)
one_hot = utils.to_categorical(padded_x)
print(one_hot) # 여기서 [0,0,0,0,0,0] 은 아무 의미가 없는 데이터를 표현하기 위해서
```

```python
x_data = one_hot[:2]
y_data = one_hot[2:]
print(x_data)
#x_data = x_data.reshape(1, x_data.shape[0], x_data.shape[1])
#y_data = y_data.reshape(1, y_data.shape[0], y_data.shape[1])
print(x_data.shape, y_data.shape)
```

```python
from tensorflow.keras.layers import Bidirectional

model = Sequential()
model.add(Input(shape=(3, 9))) # None (데이터 개수), 3, 9 
model.add(Bidirectional(LSTM(18, return_sequences=True))) # true: 모든 cell의 출력을 다 사용하겠다. 
model.add(Bidirectional(LSTM(36, return_sequences=True))) # 2장의 LSTM 깔고. 왼쪽한번, 오른쪽한번
# 나는 -> 당신을 -> 사랑합니다 도 train 하고
# 사랑합니다 -> 당신을 -> 나는 도 train 하고
model.add(Dense(18, activation='relu')) 
model.add(Dense(9, activation='sigmoid')) 

model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary() # 6개의 cell이 각각 12개씩 내보낸다
```

```python
history = model.fit(x_data, y_data, epochs=1000)
```

```python
import matplotlib.pyplot as plt
# y_vloss = history.history['val_loss']
y_loss = history.history['loss']
y_acc = history.history['accuracy']
plt.plot(np.arange(len(y_loss)), y_loss, c="red", label="loss")
plt.plot(np.arange(len(y_acc)), y_acc, c="blue", label="accuracy")
plt.legend(loc="best")
plt.show()
# train data의 loss는 항상 쭉 떨어져야 함
# test data의 loss는 떨어지다가 어느순간 올라야 함 (안 그러면 오버피팅)
# test data의 loss가 떨어지는 구간이 좀 더 늦게 나타났으면 좋겠다 (갭을 줄이고 싶다)
# 오버피팅 원인 1: train 데이터가 test 데이터의 범위를 충분히 반영하지 못함
# 오버피팅 원인 2: 모델의 복잡도가 너무 낮음 --> 모델 복잡도를 높아야 함
# 이걸 검증하려면 train, test data set을 계속 바꿔봐라. 
# 그래프가 계속 다르게 나온다면 데이터의 다양성이 너무 큰거
# 그래서 충분히 많은 양의 데이터가 필요함
# 왜 데이터가 부족한지 시각적으로 증거를 보여야 함. 
# 이 set에 대해 outlier, varience를 충분히 파악할 데이터가 모이지 않았다
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train, _), (x_test, _) = mnist.load_data() # 1차원 vector처럼 저장되어 있음
```

```python
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1).astype("float32") / 255 # 정규화 RGB는 흑백. 0~1값
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1).astype("float32") / 255 # 정규화
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Flatten, Reshape
```

```python
autoencoder = Sequential() #이미지를 뽑아내는 autoencoder

autoencoder.add(Input(shape=(28,28,1)))
autoencoder.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
autoencoder.add(MaxPooling2D(pool_size=2, padding='same'))
autoencoder.add(Conv2D(8, kernel_size=3, padding='same', activation='relu')) 
autoencoder.add(MaxPooling2D(pool_size=2, padding='same')) 
# 위는 encoder, 아래는 decoder
# 이미지를 7x7로 줄이고 feature도 4로 줄여
# 이 (7,7,4) 가 (28,82,1) 이미지의 핵심 패턴이 됨
autoencoder.add(Conv2D(4, kernel_size=3, padding='same', activation='relu')) # 모든 mnist를 7,7,4로 압축
autoencoder.add(UpSampling2D()) #이미지를 2배로 키우는데 더블로
autoencoder.add(Conv2D(8, kernel_size=3, padding='same', activation='relu'))
autoencoder.add(UpSampling2D()) 
autoencoder.add(Conv2D(16, kernel_size=3, padding='same', activation='relu'))
autoencoder.add(Conv2D(1, kernel_size=3, padding='same', activation='sigmoid'))

#입력값과 출력값이 같아야 함

autoencoder.summary()
```

```python
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
autoencoder.fit(x_train, x_train, validation_data=(x_test, x_test), batch_size=256, epochs=100)
```

```python
random_test = np.random.random.radin(x_test.shape[0], size=5)
ae_imgs = autoencoder.predict(x_test)

plt.figure(figsize=(7,2))
for i, image_idx in enumerate(random_test):
    ax = plt.subplot(2,7,i+1)
    plt.imshow(x_test[image_idx].reshape(28,28))
    ax.axis('off')
    ax = plt.subplot(2,7,7+i+1)
    plt.imshow(ae_imgs[image_idx].reshape(28,28))
    ax.axis('off')
plt.show()
```
