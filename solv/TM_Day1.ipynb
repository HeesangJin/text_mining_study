```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./dataset/sonar.csv", header=None)
x = np.array(df.values[:,0:-1], dtype='float64')
y = df.values[:,-1]
```

```python
np.unique(y, return_counts=True)
```

```python
enc = LabelEncoder()
enc.fit(y)
y_lab = enc.transform(y)
np.unique(y_lab, return_counts=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y_lab, stratify=y_lab,
                                                    test_size=0.3, 
                                                    random_state=777)
print(x_train.shape, y_train.shape)
print(np.unique(y_train, return_counts=True))
print(x_train[:2,:])
```

```python
model = Sequential()
model.add(Dense(10, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

early_stopping_cb = EarlyStopping(monitor='val_loss', patience=10)
check_pt = ModelCheckpoint(filepath="./sonar_{epoch:03d}-{val_loss:.4f}.hdf5",
                           monitor='val_loss', verbose=1, save_best_only=True)
model.fit(x_train, y_train, validation_data=(x_test,y_test), epochs=1000, 
          callbacks=[early_stopping_cb, check_pt])
```

```python
from tensorflow.keras.models import load_model

ld_model = load_model("./sonar_246-0.1990.hdf5")
```

```python
ld_model.summary()
```

```python
ld_model.evaluate(x_test, y_test)
```
